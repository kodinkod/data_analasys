# Метрика

Перед решением любой задачи важно понимать, как будет оцениваться ваше решение. В данном случае мы используем стандартную для задачи классификации метрику ROC-AUC. Ее можно вычислить, используя только предсказанные вероятности и истинные классы без конкретного порога классификации + она раотает даже если классы в данных сильно несбалансированны (примеров одного класса в десятки раз больше примеров длугого). Именно поэтому она очень удобна для соревнований.

Посчитать ее легко:


```python
from sklearn.metrics import roc_auc_score

y_true = [
    0,
    1,
    1,
    0,
    1
]

y_predictions = [
    0.1,
    0.9,
    0.4,
    0.6,
    0.61
]

roc_auc_score(y_true, y_predictions)
```

```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.compose import make_column_transformer
```

```python
train_x = data.drop(target_col, axis=1)
train_y = data[target_col]

preprocessing = make_column_transformer(
    (make_pipeline(StandardScaler(), RobustScaler()), num_cols),
    (OneHotEncoder(), cat_cols)
)

model = make_pipeline(
    preprocessing,
    LogisticRegression(max_iter=1000)
)
```

```python
param_grid = {
    'logisticregression__C': [200,150,110,100, 10, 1, 0.1, 0.01, 0.001],
    'logisticregression__solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
    'logisticregression__max_iter': [250, 500, 1000, 1500],
    'logisticregression__penalty': ['l1', 'l2']
}

search = GridSearchCV(model, param_grid, n_jobs=-1, cv=6, refit=True, scoring='roc_auc')

# запустим поиск
search.fit(train_x, train_y)

# выведем наилучшие параметры
print(search.best_params_)
```

```python
# выведем наилучшие параметры
print('Лучшие параметры: ', search.best_params_)
print('Лучший score', search.best_score_)
```

```python

predict = search.best_estimator_.predict(train_x)
roc_auc_score(train_y, predict)
```

## Применение градиентного бустинга (2 балла)

Если вы хотите получить баллы за точный ответ, то стоит попробовать градиентный бустинг. Часто градиентный бустинг с параметрами по умолчанию даст вам 80% результата за 0% усилий.

Мы будем использовать catboost, поэтому нам не надо кодировать категориальные признаки. Catboost сделает это сам (в .fit() надо передать cat_features=cat_cols). А численные признаки нормировать для моделей, основанных на деревьях не нужно.

1) Разделите выборку на train/valid. Протестируйте catboost cо стандартными параметрами.

2) Протестируйте разные занчения параметроа количества деревьев и learning_rate'а и выберите лучшую по метрике ROC-AUC комбинацию. 

(Дополнительно) Есть некоторые сложности с тем, чтобы использовать CatBoostClassifier вместе с GridSearchCV, поэтому мы не просим использовать кроссвалидацию. Но можете попробовать)

```python
from catboost import CatBoostClassifier
```

```python
train_x, valid_x, train_y, valid_y = train_test_split(train_x, train_y, test_size=0.2, random_state=42)
```

```python
boosting_model = CatBoostClassifier(eval_metric='AUC',
                                    cat_features=cat_cols,
                                    verbose = 100            #Use the Verbose logging level and set the logging period to the value of this parameter.
                                    )

boosting_model.fit(train_x, train_y,
                       cat_features=cat_cols,
                       eval_set=(valid_x, valid_y),
                       verbose=100
                   )
```

```python

boosting_model_2 = CatBoostClassifier(
    eval_metric='AUC',
    random_seed=42,
    verbose=1
)

params = {
    'num_trees': [50, 100, 150, 250, 400],
    'learning_rate': [0.03, 0.01, 0.1, 0.3, 0.005],
    'depth': [4, 6, 8, 9],
}

# Создадим GridSearchCV
cat_grid = GridSearchCV(boosting_model_2, params, scoring='roc_auc', n_jobs=-1)

# Обучим модель
cat_grid.fit(
    train_x, train_y,
    cat_features=cat_cols,
    eval_set=(valid_x, valid_y),
    verbose=False # слишком много выводит модель
)
```

```python
print('param:', cat_grid.best_params_)
print('score:', cat_grid.best_score_)
```
